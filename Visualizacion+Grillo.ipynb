Cargar el archivo "HRDataset_v14.csv" en Python. Realizar estadísticas descriptivas básicas

# 2

Realizar un histograma con los salarios. Qué rango(s) de salarios son los más populares?

```python
#Creamos la figura, le seteamos el tamaño del gráfico, tamaño de la fuente de los ejes, 
#tamaño de la fuente de los valores en los ejes, limites de los ejes (entre 0 y salario máximo).
#Seteamos el gráfico y su título

plt.figure()

plt.rc('figure', figsize=(20,10))

plt.rc('axes', titlesize=30, labelsize=24)

plt.tick_params(labelsize=16)

plt.xlim(0, df_HR['Salary'].max())

ax = sns.histplot(data=df_HR, x='Salary', binwidth=5000)

ax.set(title='Histograma de salarios')

plt.show()
```

### Dado que cada barra representa un rango de 5000 de salario. Se observa que el rango de salario más popular es entre 60000 - 65000

```python
#Visualización de la distribución de los salarios incluyendo la distinción por sexo

plt.figure()

plt.tick_params(labelsize=16)

plt.xlim(0, df_HR['Salary'].max())

ax = sns.histplot(data= df_HR, x= "Salary", hue="Sex", multiple="stack", binwidth=5000)

ax.set(title='Histograma de salarios con distincion de sexo')

plt.show()
```

# 3

Realizar un gráfico de violín con los salarios, discriminados por género. 

```python
#Gráfico de violin de los salarios, con discriminación por género.

plt.figure()

plt.rc('figure', figsize=(8,8))

plt.rc('axes', titlesize=18, labelsize=16)

ax = sns.violinplot(data= df_HR, x='Sex' , y= 'Salary', palette="Set3")

ax.set(title='Gráfico de violín de salarios con distincion de sexo')

plt.show()
```

```python
#Gráfico de violín de salarios, agrupados por raza de la persona, discriminado por género

plt.figure()

ax = sns.violinplot(data=df_HR, x='Salary', y='RaceDesc', hue='Sex', split=True, palette='Set3')

ax.set(title='Gráfico de violín de salarios con distincion de sexo, por raza')

ax.set_ylabel('Raza de la persona')

plt.show()
```

# 4

Graficar la serie de tiempo correspondiente a la fecha de contratación (DateofHire)

```python
#Creo el df de las contrataciones por fecha. Creo el gráfico, y configuro nombramientos de ejes y título

fig, ax = plt.subplots(figsize=(20,5))

df_contrataciones_por_fecha = pd.DataFrame(df_HR['DateofHire'].value_counts()).sort_index()

ax.plot(df_contrataciones_por_fecha.index, df_contrataciones_por_fecha['DateofHire'], label='Contrataciones')

ax.set_xlabel('Fecha')

ax.set_ylabel('Contrataciones')

ax.set_title('Contrataciones por fecha')

plt.xticks(rotation=90)

ax.legend()
```

```python
#Al verse de forma confusa, convierto la columna a datetime para que lo grafique de una forma agradable y entendible
#Seteo la columna de Fecha de contratacion como Datetime
#Calculo las columnas mes.año y año, para también graficar por estas dimensiones

df_HR['DateofHire'] = pd.to_datetime(df_HR['DateofHire'], infer_datetime_format=True)


df_HR['YearofHire'] = df_HR['DateofHire'].dt.year

df_HR['Month-YearofHire'] = df_HR['DateofHire'].dt.strftime('%Y.%m')

df_HR.head()
```

```python

fig, ax = plt.subplots(figsize=(20,5))

df_contrataciones_por_fecha = pd.DataFrame(df_HR['DateofHire'].value_counts()).sort_index()

ax.plot(df_contrataciones_por_fecha.index, df_contrataciones_por_fecha['DateofHire'], label='Contrataciones')

ax.set_xlabel('Fecha')

ax.set_ylabel('Contrataciones')

ax.set_title('Contrataciones por fecha')

ax.legend()
```

```python
#Generación de gráficos de diferentes períodos

#seteo general del gráfico
fig , ax = plt.subplots(nrows=3 , ncols=1 ,figsize=(18,24), sharex=False , sharey=False )


#DataFrames por dimensión de eje X

df_contrataciones_por_fecha = pd.DataFrame(df_HR['DateofHire'].value_counts()).sort_index()

df_contrataciones_por_mes_año = pd.DataFrame(df_HR['Month-YearofHire'].value_counts()).sort_index()

df_contrataciones_por_año = pd.DataFrame(df_HR['YearofHire'].value_counts()).sort_index()


#Creación de gráficos
ax[0].plot(df_contrataciones_por_año.index, df_contrataciones_por_año['YearofHire'],
           label='Contrataciones por año')

ax[1].plot(df_contrataciones_por_mes_año.index, df_contrataciones_por_mes_año['Month-YearofHire'],
           label='Contrataciones por mes-año')

ax[2].plot(df_contrataciones_por_fecha.index, df_contrataciones_por_fecha['DateofHire'], label='Contrataciones por fecha')



for tick in ax[1].get_xticklabels():
    tick.set_rotation(80)
    
    
ax[0].set_ylabel('Contrataciones')
ax[1].set_ylabel('Contrataciones')
ax[2].set_ylabel('Contrataciones')

ax[0].set_xlabel('Año')
ax[1].set_xlabel('Año.Mes')
ax[2].set_xlabel('Fecha')


ax[0].set_title('Contrataciones por Períodos')
```
