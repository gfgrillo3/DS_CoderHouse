# 1

Con el archivo "Summer-Olympic-medals-1976-to-2008.csv", construir un objeto de series de tiempo con el índice igual al año de los juegos olímpicos

```python
df_olympics = pd.read_csv("C:\\Users\\usuario\\Desktop\\Data Science\\Summer-Olympic-medals-1976-to-2008.csv", sep=',')
```

```python
#Filas completamente vacías. Las elimino. Convierto el Año de float a int, y luego lo seteo como índice
#Al no trabajar el año posteriormente, no hace falta convertirlo a datetime

df_olympics.dropna(inplace=True)

df_olympics['Year'] = pd.to_numeric(df_olympics['Year'], errors='coerce')

df_olympics['Year'] = df_olympics['Year'].astype(np.int64)

df_olympics.set_index('Year', inplace=True)
```

Separar nombres de apellidos en dos columnas distintas usando la coma como separador

```python
df_olympics[['Last_name', 'First_names']] = df_olympics['Athlete'].str.split(',', expand=True, n=1)
```

# 3

Obtener medidas resumen del conjunto de datos. Cuál es el país que ganó más medallas?

```python
df_olympics.describe()
```

```python
#Columnas, cantidad de regisros totales, comienzo y fin de índices, cantidad de registros no nulos, y tipos de datos

df_olympics.info()
```

```python
#Cantidad de nulos por columna

df_olympics.isnull().sum()
```

```python
#tabla con el total de medallas por pais a lo largo de la historia de los juegos olimpicos, ordenado de mayor cantidad
#de medallas a menor

df_maximas_medallas_totales = df_olympics.groupby('Country').agg({'Medal':'count'}).sort_values('Medal', ascending=False)

df_maximas_medallas_totales
```

```python
print('{} es el país con mas medallas a lo largo de todos los juegos olímpicos, con un total de {:d} medallas'.format(
     df_maximas_medallas_totales.idxmax()[0],
     df_maximas_medallas_totales.max()[0]
     ))
```

# 4

Construir una tabla que muestre cuántas medallas obtuvieron los hombres en total en cada año que se realizó el evento.

```python
#Medallas totales por olimpiadas de atletas masculinos

df_medallas_masculinas = df_olympics[df_olympics.Gender == "Men"].groupby('Year').agg({'Medal':'count'})

df_medallas_masculinas.rename(columns={"Medal": "Medals_Count"}, inplace=True)

df_medallas_masculinas
```

```python
#Medallas totales por olimpiadas de atletas masculinos
#Medallas contabilizadas por el tipo de medalla


df_medallas_masculinas_por_tipo = df_olympics[df_olympics.Gender == "Men"].groupby(['Year', 'Medal']).agg({'Medal':'count'})

df_medallas_masculinas_por_tipo.rename(columns={"Medal": "Medals_Qty"}, inplace=True)

df_medallas_masculinas_por_tipo.reset_index(inplace=True)

df_medallas_masculinas_por_tipo = df_medallas_masculinas_por_tipo.pivot(index='Year', columns='Medal', values='Medals_Qty')
```
