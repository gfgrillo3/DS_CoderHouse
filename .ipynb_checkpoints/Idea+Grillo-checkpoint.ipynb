Leemos el json y vemos como viene la info

```python
#usamos la funci√≥n de pandas que nos permite leer un json directo de una URL

df = pd.read_json("https://jsonplaceholder.typicode.com/users")

df
```

Vemos que las columnas "address" y "company" siguen siendo jsons, por lo que era un json "anidado"

Podemos normalizar el JSON anidado de 2 formas: 
    
    - Separando estas columnas en diferentes dataframes (quedar√≠an tablas simil a BBDD relacional)
    
    - Aperturando estas columnas en el mismo dataframe

# 

### Separemos columnas en diferentes df

Creemos una copia del df que leimos para no pisar la info

```python
df_users = df.copy()
```

Creamos el df de direcciones y el df de compa√±ias con las columnas que necesitariamos en un principio

```python
df_address = pd.DataFrame(columns=['id_user', 'address_id', 'address'])
df_companies = pd.DataFrame(columns=['id_user', 'company_id', 'company'])

df_companies.head()
```

Agreguemos a ambos DFs los ids de los usuarios

```python
df_address['id_user'] = df_users['id']
df_companies['id_user'] = df_users['id']
```

Agreguemos a los DFs creados los indices como columnas, para que sean los IDs y luego agregarlos al df original para tener la referencia de qu√© registro le pertenece a cada usuario

```python
df_address['address_id'] = df_address.index
df_companies['company_id'] = df_companies.index
```

Agreguemos a los dfs, las columnas del df que leimos para tener la columna con el json

```python
df_address['address'] = df_users['address']
df_companies['company'] = df_users['company']

df_companies.head()
```

Ahora con las columnas con la info como json en sus respectivos dataframes, pasamos a normalizarlas.

Usamos json_normalize de pandas, que nos permite normalizar un JSON en formato tabla.

Seteamos max_level=0 para que la apertura la haga en un √∫nico nivel del JSON. En caso de que esta tenga otro nivel m√°s (est√© nuevamente anidada), deber√≠amos normalizarla devuelta

Estas normalizaciones las joineamos con el mismo df para poder mantener las otras columnas que no ten√≠an informaci√≥n en formato json


```python
df_address = df_address.join(pd.json_normalize(df_address['address'], max_level=0))
df_companies = df_companies.join(pd.json_normalize(df_companies['company'], max_level=0))

df_companies.head()
```

Luego de las normalizaciones dropeamos las columnas que ten√≠an la info en formato json

```python
df_address.drop(columns='address', inplace=True)
df_companies.drop(columns='company', inplace=True)

df_companies
```

Para mantener la referencia del df original, con su respectiva direcci√≥n y compa√±√≠a, agreguemosle los ids que creamos en los nuevos DF con un Join.

Al tener diferentes nombres de columnas, tenemos que especificar sobre qu√© campo hacemos el join.

Tambi√©n borremos las columnas "id_user" que se agrega al hacer el join con los dfs creados, y address y company que ten√≠an la info como json.

```python
#Join con address y eliminamos columnas relacionadas con address

df_users = df_users.merge(df_address[['id_user','address_id']], left_on='id', right_on='id_user')

df_users.drop(columns=['id_user', 'address'], inplace=True)


#Join con company y eliminamos columnas relacionadas con address

df_users = df_users.merge(df_companies[['id_user','company_id']], left_on='id', right_on='id_user')

df_users.drop(columns=['id_user', 'company'], inplace=True)
```

Veamos como quedaron los dfs de usuarios, compa√±ias y direcciones

Los dfs de usuarios y compa√±ias ya quedaron OK, pero podemos ver que en el DF de direcciones, la columna geo sigue teniendo formato JSON.

Hagamos el mismo proceso de separar la columna GEO en un DF nuevo, y veamos como quedan los df de direcciones y de puntos geograficos

```python
df_geo_points = pd.DataFrame(columns=['address_id', 'geo_point_id', 'geo_point'])

df_geo_points['address_id'] = df_address['address_id']

df_geo_points['geo_point_id'] = df_geo_points.index

df_geo_points['geo_point'] = df_address['geo']

df_geo_points = df_geo_points.join(pd.json_normalize(df_geo_points['geo_point'], max_level=0))

df_geo_points.drop(columns='geo_point', inplace=True)

df_address = df_address.merge(df_geo_points[['address_id','geo_point_id']])

df_address.drop(columns='geo', inplace=True)
```

Ahora s√≠ ya tenemos separado el JSON original en 4 DF diferentes que se fueron desprendiendo seg√∫n las anidaciones.

- df_users, con info de los usuarios

- df_companies, con info de las compa√±ias de los usuarios

- df_address, con info de las direcciones de los usuarios

- df_geo_points, con info de los puntos geogr√°ficos de las direcciones


Las relaciones son de la siguiente forma:

df_users -> df_address -> df_geo_points

df_users -> df_companies

# 

### Aperturemos estas columnas en el mismo dataframe

Normalmente, al consumir la informaci√≥n, vamos a querer que quede toda en el mismo DF para posteriormente analizarla y utilizarla

Para esto, normalizemos el df le√≠do directamente 

```python
df_users_aggr = df.copy()
```

```python
#Normalicemos la columna company agregandole a las columnas el prefijo company_ .
#Normalicemos la columna address agregandole a las columnas el prefijo address_ . Recordemos que tiene otro JSON anidado de
#la columna "geo". Setiemos para que, al normalizar el nivel de la columna geo, separe los nombres con un '_'.
#Eliminemos las columnas con la info en JSON

df_users_aggr = df_users_aggr.join(pd.json_normalize(df_users_aggr['company']).add_prefix('company_'))

df_users_aggr = df_users_aggr.join(pd.json_normalize(df_users_aggr['address'], sep='_').add_prefix('address_'))

df_users_aggr.drop(columns=['company', 'address'], inplace=True)

df_users_aggr
```
