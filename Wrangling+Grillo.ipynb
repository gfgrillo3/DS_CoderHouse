Cuantos ID tienen en com√∫n. Nombrar cuales no tienen en com√∫n

```python
#Me quedo s√≥lo con la columna ID de ambos df, y hago un inner join para quedarme s√≥lo con ids que est√°n en ambos df
#Luego, cuento los valores √∫nicos

df_inner = df_w[['ID']].merge(df_p['ID'], how='inner', on='ID')

print("En total, tienen {:d} ID's en com√∫n".format(df_inner['ID'].nunique()))
```

```python
#Hago un outer join para quedarme con los ids compartidos y los que no se comparten, pero esta vez, agregandole
#el parametro indicator=True, que me indica en que df est√° el valor por el que se joinea (ID).
#De estos, me quedo con todos los que no sean "both", ya que representa que esta en ambos df.
#Imprimo para mostrar los ID'S que no est√°n compartidos en ambos df.

df_outer =  df_w[['ID']].merge(df_p['ID'], how='outer', on='ID', indicator=True)

df_outer = df_outer[df_outer['_merge']!='both']

df_outer[['ID']]
```

# 2

Corregir los target a solo dos valores distintos ‚Äò<=50k‚Äô y ‚Äò>50k‚Äô

```python
#Primero vemos los diferentes valores que vienen para saber como modificarlos

df_joined['Target'].unique()
```

```python
#Vemos que tenemos un nulo, asi que lo dropeamos.
#Luego vemos los signos. Si contiene el signo '<' en la cadena, le ponemos '<=50K', sino, '>50K'


df_joined.dropna(subset=['Target'], inplace=True)

df_joined['Target'] = df_joined['Target'].apply(lambda target: '<=50K' if ('<' in target) else '>50K')
```

```python
#Ahora podemos ver que tenemos s√≥lo 2 valores

df_joined['Target'].unique()
```

# 3

Calcular la proporci√≥n que hay entre personas con salarios >50k respecto del total para 
Entre personas de distinto sexo, Raza y Educacion. Por ejemplo cu√°l es el porcentaje de mujeres con un salario mayor a 50K y compararlo con el de los hombres. Lo mismo para personas de distintas razas y Educaci√≥n

```python
#Creamos un df agrupando por la columna correspondiente, para cada variable solicitada, y contando la cantidad
#de registros de las mismas (cuento ID).
#Renombro las columnas ID por personas totales.

df_aggr_sex = df_joined[['ID','Sex']].groupby('Sex').agg({'ID' : 'count'})

df_aggr_race = df_joined[['ID','Race']].groupby('Race').agg({'ID' : 'count'})

df_aggr_education = df_joined[['ID','Education']].groupby('Education').agg({'ID' : 'count'})


df_aggr_sex.rename(columns={'ID' : 'total_persons'}, inplace=True)
df_aggr_race.rename(columns={'ID' : 'total_persons'}, inplace=True)
df_aggr_education.rename(columns={'ID' : 'total_persons'}, inplace=True)
```

```python
#A cada df le agrego, por cada agrupaci√≥n, la cantidad de personas con salarios >50K.
#Utilizo la columna ID porque es sobre la que se hace el count.
#Para facilitar lectura, creo un df filtrado con target >50K.


df_greater_than_50K = df_joined[df_joined['Target']=='>50K']

df_aggr_sex['>50K'] = df_greater_than_50K[['ID','Sex']].groupby('Sex').agg({'ID' : 'count'})['ID']

df_aggr_race['>50K'] = df_greater_than_50K[['ID','Race']].groupby('Race').agg({'ID' : 'count'})['ID']

df_aggr_education['>50K'] = df_greater_than_50K[['ID','Education']].groupby('Education').agg({'ID' : 'count'})['ID']
```

```python
#Agrego una columna donde me indique el porcentaje con target >50K sobre el total de personas en el dataset.
#Esto se leer√≠a, en el caso de la agrupaci√≥n por sexo: 
#"El X% del total de personas son mujeres con target >50K, y el Y% del total de personas son hombres con target >50K".


df_aggr_sex['perc_greater_than_50K_over_total'] = df_aggr_sex['>50K'].apply(lambda qty : 
                                                                    '{:.2%}'.format(qty / df_aggr_sex['total_persons'].sum()))


df_aggr_race['perc_greater_than_50K_over_total'] = df_aggr_race['>50K'].apply(lambda qty : 
                                                                    '{:.2%}'.format(qty / df_aggr_race['total_persons'].sum()))


df_aggr_education['perc_greater_than_50K_over_total'] = df_aggr_education['>50K'].apply(lambda qty : 
                                                                    '{:.2%}'.format(qty / df_aggr_education['total_persons'].sum()))
```

```python
#Agrego una columna donde me indique el porcentaje con target >50K sobre la cantidad de personas de la clase actual.
#Esto se leer√≠a, en el caso de la agrupaci√≥n por sexo: 
#"El X% de mujeres tienen target >50K, y el Y% de hombres tienen target >50K".


df_aggr_sex['perc_greater_than_50K_of_class'] = (df_aggr_sex['>50K'] / df_aggr_sex['total_persons']) *100


df_aggr_race['perc_greater_than_50K_of_class'] = (df_aggr_race['>50K'] / df_aggr_race['total_persons']) *100


df_aggr_education['perc_greater_than_50K_of_class'] = (df_aggr_education['>50K'] / df_aggr_education['total_persons']) *100
```

# 

### VEMOS LOS RESULTADOS 

```python
#Quiero ver la informaci√≥n del df de los sexos, ordenando los sexos, seg√∫n la cantidad de personas del sexo con target
#>50K, de mayor a menor

df_aggr_sex.sort_values('perc_greater_than_50K_of_class', ascending=False)
```

Podemos ver que el 30% de los hombres tienen target >50K, mientras que las mujeres solamente el 10% tiene target >50K

# 

```python
#Quiero ver la informaci√≥n del df de las razas, ordenando los razas, seg√∫n la cantidad de personas de la raza con target
#>50K, de mayor a menor

df_aggr_race.sort_values('perc_greater_than_50K_of_class', ascending=False)
```

Podemos ver que las personas de raza Asian-Pac-Islander son las que, dentro de su raza, tienen mayor porcentaje de personas con target >50K con casi un 27%

En lo opuesto, las personas de raza Amer-Indian-Eskimo son las que, dentro de su raza, tienen menor porcentaje de personas con target >50K con un 11%

Sin embargo, si analizamos, por raza, el porcentaje de personas con target >50K respecto al TOTAL de personas, notamos que la raza White tiene el mayor porcentaje de personas con target >50K respecto al total de personas, con un 21%

# 

```python
#Quiero ver la informaci√≥n del df de educaci√≥n, ordenando los niveles de educaci√≥n, seg√∫n la cantidad de personas del
#nivel de educaci√≥n con target >50K, de mayor a menor

df_aggr_education.sort_values('perc_greater_than_50K_of_class', ascending=False)
```

Podemos ver que las personas con educaci√≥n Prof-school son las que, dentro de su nivel de educaci√≥n, tienen mayor porcentaje de personas con target >50K, con casi un 74%.

En lo opuesto, las personas con educaci√≥n Preschool son las que, dentro de su nivel de educaci√≥n, tienen menor porcentaje de personas con target >50K, con un 1%.

Sin embargo, si analizamos, por nivel de educaci√≥n, el porcentaje de personas con target >50K respecto al TOTAL de personas, notamos que el nivel de educaci√≥n Masters tiene el mayor porcentaje de personas con target >50K respecto al total de personas, con un 3%.
