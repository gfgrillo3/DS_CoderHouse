# 3

Calcular la proporción que hay entre personas con salarios >50k respecto del total para 
Entre personas de distinto sexo, Raza y Educacion. Por ejemplo cuál es el porcentaje de mujeres con un salario mayor a 50K y compararlo con el de los hombres. Lo mismo para personas de distintas razas y Educación

```python
#Creamos un df agrupando por la columna correspondiente, para cada variable solicitada, y contando la cantidad
#de registros de las mismas (cuento ID).
#Renombro las columnas ID por personas totales.

df_aggr_sex = df_joined[['ID','Sex']].groupby('Sex').agg({'ID' : 'count'})

df_aggr_race = df_joined[['ID','Race']].groupby('Race').agg({'ID' : 'count'})

df_aggr_education = df_joined[['ID','Education']].groupby('Education').agg({'ID' : 'count'})


df_aggr_sex.rename(columns={'ID' : 'total_persons'}, inplace=True)
df_aggr_race.rename(columns={'ID' : 'total_persons'}, inplace=True)
df_aggr_education.rename(columns={'ID' : 'total_persons'}, inplace=True)
```

```python
#A cada df le agrego, por cada agrupación, la cantidad de personas con salarios >50K.
#Utilizo la columna ID porque es sobre la que se hace el count.
#Para facilitar lectura, creo un df filtrado con target >50K.


df_greater_than_50K = df_joined[df_joined['Target']=='>50K']

df_aggr_sex['>50K'] = df_greater_than_50K[['ID','Sex']].groupby('Sex').agg({'ID' : 'count'})['ID']

df_aggr_race['>50K'] = df_greater_than_50K[['ID','Race']].groupby('Race').agg({'ID' : 'count'})['ID']

df_aggr_education['>50K'] = df_greater_than_50K[['ID','Education']].groupby('Education').agg({'ID' : 'count'})['ID']
```

```python
#Agrego una columna donde me indique el porcentaje con target >50K sobre el total de personas en el dataset.
#Esto se leería, en el caso de la agrupación por sexo: 
#"El X% del total de personas son mujeres con target >50K, y el Y% del total de personas son hombres con target >50K".


df_aggr_sex['perc_greater_than_50K_over_total'] = df_aggr_sex['>50K'].apply(lambda qty : 
                                                                    '{:.2%}'.format(qty / df_aggr_sex['total_persons'].sum()))


df_aggr_race['perc_greater_than_50K_over_total'] = df_aggr_race['>50K'].apply(lambda qty : 
                                                                    '{:.2%}'.format(qty / df_aggr_race['total_persons'].sum()))


df_aggr_education['perc_greater_than_50K_over_total'] = df_aggr_education['>50K'].apply(lambda qty : 
                                                                    '{:.2%}'.format(qty / df_aggr_education['total_persons'].sum()))
```

```python
#Agrego una columna donde me indique el porcentaje con target >50K sobre la cantidad de personas de la clase actual.
#Esto se leería, en el caso de la agrupación por sexo: 
#"El X% de mujeres tienen target >50K, y el Y% de hombres tienen target >50K".


df_aggr_sex['perc_greater_than_50K_of_class'] = (df_aggr_sex['>50K'] / df_aggr_sex['total_persons']) *100


df_aggr_race['perc_greater_than_50K_of_class'] = (df_aggr_race['>50K'] / df_aggr_race['total_persons']) *100


df_aggr_education['perc_greater_than_50K_of_class'] = (df_aggr_education['>50K'] / df_aggr_education['total_persons']) *100
```
